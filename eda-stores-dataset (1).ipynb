```python
import matplotlib.pyplot as plt
import seaborn as sns
```

## Let us look at some graphs of the columns

```python
fig,axes = plt.subplots(2,2,figsize=(10,10))
sns.histplot(data = data, x = 'Store_Area',binwidth = 10,kde=True,ax=axes[0,0])
sns.histplot(data = data, x = 'Items_Available',binwidth = 50,kde=True,ax=axes[0,1])
sns.histplot(data = data, x = 'Daily_Customer_Count',binwidth = 10,kde=True,ax=axes[1,0])
sns.histplot(data = data, x = 'Store_Sales',binwidth = 1000,kde=True,ax=axes[1,1])
```

## We see that the histogram plots follow an approximate normal distribution.

### Let us check for some correlation among the columns.

```python
sns.heatmap(data.corr(),cmap="YlGnBu",linewidths = 0.5,linecolor = 'red' )
```

```python
sns.scatterplot(x=data['Store_Area'],y=data['Items_Available'])
```

## We see a strong correlation between Store Area and Items Availble, which is expected. The larger the available space , more the products.

## Lets analyze the stores with the highest sales.

```python
high_sales = data.sort_values(by='Store_Sales',ascending=False).head(100)
high_sales
```

```python
high_sales.describe()
```

```python
fig,axes = plt.subplots(1,2,figsize=(10,5))
sns.scatterplot(x=high_sales['Daily_Customer_Count'],y=high_sales['Store_Sales'])
sns.scatterplot(x=high_sales['Items_Available'],y=high_sales['Store_Sales'],ax=axes[0])
```

## No visible Trends found with respect to the sales of the stores.

```python
sns.histplot(x=high_sales['Daily_Customer_Count'],y=high_sales['Store_Sales'],cbar=True,log_scale=(False,True))
```

## We see that there exists a lot of stores with high Customer Count but low Sales.
